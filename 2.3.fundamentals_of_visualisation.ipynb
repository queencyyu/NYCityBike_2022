### 2.3 Fundamentals of Visualisation - Matplotlib

```python
# Import Libraries

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
```

```python
# Import data

df = pd.read_csv('nycbike_data.csv', index_col = 0)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

#### Create a line plot of the temperatures for 2022 by using pandas plotting function

```python
# Set the date column to be the index of df_temp

df_temp = df.set_index('date')
```

```python
df_temp.head()
```

```python
# Set agg.path.chunksize to a larger value

plt.rcParams['agg.path.chunksize'] = 10000
```

```python
df_temp = df_temp.sort_values(by='date')
```

```python
df_temp.head()
```

```python
# Create line plot using pandas

df_temp['avgTemp'].plot()
plt.show()
```

#### Create a column for count of trips per day and merged with the dataframe

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
# Merge both dataframe

df_group.rename(columns = {'ride_id':'trips_per_day'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = True)
```

```python
# Checking the merge

print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 
```

```python
df_temp = df.set_index('date')

df_temp.head()
```

#### Create a line chart of trips per day and temperatures plotted on dual axis

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(12, 6))

# Make a plot of the bike trips per day
ax.plot(df_temp['trips_per_day'],color = "darkblue")
ax.set_xlabel("Year 2020", fontsize = 12) # set x-axis label
ax.set_ylabel("Bike trips count",color = "darkblue", fontsize = 12) # set y-axis label

# Create the dual axis
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["avgTemp"], color = "sandybrown")
ax2.set_ylabel("Average temperatures", color = "sandybrown",fontsize=12)
plt.title('Bike Trips count and avg Temperatures in 2020', fontsize = 14)
plt.show()
```

##### To achieve above results, I have used the object-oriented approach, to assign specific code for each plots. Firstly, i have used the objects fig and ax as subplots to the charts as there will be 2 y-axes that i need to work with. Then i started to create the plot for the bike trips count first (set the label name, colour, etc.) Afterwards, i use twinx() function to have the matplotlib put the second axes in the same plot, to make it a dual axis plot. Finally, i worked on the second plot for the average temperatures.

```python
# save the plot as a file
fig.savefig('trips_temperatures_dual_axis.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
